# Replace Cells with New Values

```python
import pandas as pd
df = pd.read_excel('Energy Indicators.xls')
df = df.iloc[16:243]
df.head()
```

```python
df.tail(30)
```

```python
df['Environmental Indicators: Energy'] = df['Environmental Indicators: Energy'].str.replace('\d+', '')

df['Environmental Indicators: Energy'].replace({"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"}, inplace = True)
```
