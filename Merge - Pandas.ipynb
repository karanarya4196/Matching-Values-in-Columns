# Merge - Pandas

```python
import pandas as pd
from functools import reduce
```

```python
df = pd.read_csv('fifadata.csv')
```

```python
df.head()
```

## Make a Subset Dataframe

```python
nationality_vs_age = pd.crosstab(df['Nationality'], df['Age'])
nationality_vs_age['Total'] = nationality_vs_age.iloc[:, :].sum(axis=1)
nationality_vs_age
```

## Making a Series using Groupby

```python
nationality_vs_potential = df.groupby('Nationality')['Potential'].mean()
nationality_vs_potential
```

# Merging a Dataframe and Series

```python
merged_nationality_age_potential = nationality_vs_age.merge(pd.DataFrame(nationality_vs_potential), on = 'Nationality')
merged_nationality_age_potential
```
