# Arithmetic Operations on a Column

```python
import pandas as pd
```

```python
df = pd.read_csv('fifadata.csv')
```

```python
df.head()
```

## Addition

```python
df['Add_Age'] = df['Age'] + 5
df['Add_Age']
```

## Subtraction

```python
df['Sub_Age'] = df['Age'] - 5
df['Sub_Age']
```

## Multiplication

```python
df['Mult_Age'] = df['Age'] * 5
df['Mult_Age']
```

## Division

```python
df['Div_Age'] = df['Age']/2
df['Div_Age']
```
