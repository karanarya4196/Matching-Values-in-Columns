# Matching Values in Columns

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('wiki_movie_plots_deduped.csv')
```

```python
df.head()
```

### The follwing snippet shows the method to match the column to a particular value such as 'comedy'.

`comedy_genre` is a data frame.

```python
comedy_genre = df[df['Genre'] == "comedy"]
comedy_genre
```

### What if we wanna know the distribution of comedy movies across all origins or ethnicities?

```python
comedy_genre['Origin/Ethnicity'].value_counts()
```

```python
Looks like the most number of comedy movies are from American origin and the least are of Bangladeshi/Maldivian origin.
```

### Creative thoughts in mind? Plot curves to better visualise and understand the above results!

```python
comedy_genre['Origin/Ethnicity'].value_counts().plot(kind = 'bar')
```

### Counting the number of comedy movies in the file

There are a total of 4279 comedy movies in the database.

```python
comedy_genre.count()
```

### Creating an array of Boolean values

```python
array_comedy_genre = df['Genre'] == "comedy"
array_comedy_genre
```

### The following code counts the number of comedy movies from Bollywood.

```python
is_comedy = df['Genre'] == "comedy"
in_bollywood = df['Origin/Ethnicity'] == "Bollywood"
df[is_comedy & in_bollywood][:5]
```

```python
df[is_comedy & in_bollywood].count()
```

Hence, we find that there are a total of 4379 comedy movies out of which 154 are from Bollywood.
