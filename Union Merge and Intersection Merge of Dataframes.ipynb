# Union Merge and Intersection Merge of Dataframes

```python
import pandas as pd
fifa = pd.read_csv('fifadata.csv')
energy = pd.read_excel('Energy Indicators.xls')
```

```python
energy = energy.iloc[16:243]
energy.drop(energy.columns[[0,1]], axis = 1, inplace = True)
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy.head()
```

```python
fifa.head()
```

```python
fifa.rename(index = str, columns = {'Nationality': 'Country'}, inplace = True)
fifa.head()
```

## Union Merge

```python
union_merge = fifa.merge(energy, on = 'Country', how = 'outer')
union_merge
```

18317 rows × 92 columns

## Intersection Merge

```python
intersection_merge = fifa.merge(energy, on = 'Country', how = 'inner')
intersection_merge
```

8775 rows × 92 columns
