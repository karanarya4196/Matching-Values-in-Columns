# Mapping using Dictionary

```python
import pandas as pd
```

```python
df = pd.read_csv('fifadata.csv')
df.head()
```

```python
df.rename(index = str, columns = {'Nationality': 'Country'}, inplace = True)
```

```python
country_list = ['Argentina', 'Portugal', 'Japan', 'Belgium', 'Germany', 'Brazil', 'France']
df = df[df['Country'].apply(lambda x: x in country_list)]
```

```python
df['Country'].unique()
```

```python
df.info()
```

```python
df.set_index('Country', inplace = True)
df.head()
```

```python
ContinentDict  = {'Argentina':'South America', 
                  'Belgium':'Europe', 
                  'Japan':'Asia', 
                  'Portugal':'Europe', 
                  'Brazil':'South America', 
                  'Germany':'Europe', 
                  'France':'Europe'}

df.rename(index = ContinentDict, inplace = True)
df.head()
```
